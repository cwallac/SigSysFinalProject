```python
import matplotlib.pyplot as plt
import csv
import thinkplot
import thinkdsp
import numpy as np
with open('b247.csv', 'rb') as f:
    reader = csv.reader(f)
    your_list = list(reader)
your_list = [float(x[0]) for x in your_list]
wave = thinkdsp.Wave(np.asarray(your_list),framerate=833)
wave.plot()
thinkplot.Config(title="240Hz guitar strum",xlabel="Time(seconds)",ylabel="Amplitude(Volts)")

```

```python
lags, corrs = autocorr(wave)
thinkplot.plot(lags, corrs)
thinkplot.config(title="Autocorrelation of 240Hz Strum",xlabel='lag (index)', ylabel='correlation', ylim=[-1, 1],xlim=[1,100])
```

```python
def corrcoef(xs, ys):
    return np.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```
